```python
default = 'default-971039e0-830c-11e9-9e0b-02c5b84a8036'
response = client.list_jobs(jobQueue=default, jobStatus='RUNNABLE')
response
    
```

```python
for job in response['jobSummaryList']:
    client.terminate_job(jobId=job['jobId'], reason='Too many resources - olga killing')
```

```python
highpriority = 'highpriority-971039e0-830c-11e9-9e0b-02c5b84a8036'
response = client.list_jobs(jobQueue=highpriority, jobStatus='RUNNABLE')
for job in response['jobSummaryList']:
    if 'sourmash' in job['jobName']:
        client.terminate_job(jobId=job['jobId'], reason='Too many resources - olga killing')
```
