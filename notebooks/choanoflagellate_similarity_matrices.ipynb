```python
similarity = pd.read_csv(filename)
# similarity = similarity.replace(1, np.nan)
similarity.head(20)
```

```python
sns.clustermap(similarity, #annot=True, fmt='.3f', annot_kws={"size": 8}, 
               vmax=0.2, 
               method='ward', metric='cosine')
```

```python
from skbio import DistanceMatrix
from skbio.tree import nj
```

```python
dm = DistanceMatrix(1-similarity.values, similarity.columns)
dm
```

```python
tree = nj(dm)
print(tree.ascii_art())
```
